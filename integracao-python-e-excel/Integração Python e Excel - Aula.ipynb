##Integração Python e Excel

```python
import pandas as pd
```

```python
base_vendas = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Integração Excel - Python Pandas/BaseVendas.xlsx', skiprows = 1, sheet_name = 'Vendas')
base_vendas.head(5)
```

```python
base_vendas = base_vendas.drop('Unnamed: 0', axis = 1)
base_vendas
```

```python
base_vendas.columns
```

```python
base_vendas[['Data', 'Quantidade', 'ID do Produto', 'Valor Total da Compra']]
```

```python
base_produtos = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Integração Excel - Python Pandas/BaseProdutos.xlsx', skiprows = 1)
base_produtos.drop('Unnamed: 0', axis = 1, inplace = True)
base_produtos
```

```python
base_lojas = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Integração Excel - Python Pandas/BaseLojas.xlsx', skiprows = 1)
base_lojas.drop('Unnamed: 0', axis = 1, inplace = True)
base_lojas
```

```python
base_compradores = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Integração Excel - Python Pandas/BaseCompradores.xlsx', skiprows = 1)
base_compradores.drop(['Unnamed: 0', 'Column1'], axis = 1, inplace = True)
base_compradores.tail(5)
```

```python
base_compradores = base_compradores[base_compradores['ID do Comprador'].notna()]
base_compradores.tail(5)
```

```python
base_completa = base_vendas.merge(base_produtos, on = 'ID do Produto', how = 'left') # junção base vendas e base produtos
base_completa.head()
```

```python
base_compradores = base_compradores[['ID do Comprador','Nome Completo', 'Idade']]
base_completa = base_completa.merge(base_compradores, on = 'ID do Comprador', how = 'left')
base_completa.head()
```

```python
base_lojas = base_lojas[['ID da Loja', 'Cidade', 'Estado']]
```

```python
base_completa = base_completa.merge(base_lojas, on = 'ID da Loja', how = 'left')
base_completa.head()
```
